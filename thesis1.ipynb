```python
class Person:
    
    def create_cells(self):
        self.cell_list=np.zeros((self.cell_count, self.fcount))
        
        
        for i in range (0, self.fcount):
            self.cell_list[:,i ]= np.random.normal(self.m[i], self.sigm, self.cell_count)
    
    
    def __init__(self, t):
        
        
        if t=='rmg':
            self.m=np.array([x for x in range (0, 31) if x%2==0])
            print(self.m)

        if t=='fam':
            self.m=np.array([x for x in range (0,31) if x%2==1])
            
            print(self.m)
            
        self.cell_count=np.random.randint(10, 30)
        self.fcount=15
        #set variace
        self.sigm=1
        
        self.create_cells()
        self.mean=np.mean(self.cell_list)
    
    def get_mean(self, t):
        
        return np.mean(self.cell_list, axis=0)[t]
    
    def get_min(self, t):
        
        return min(self.cell_list[:, t])
    
    def get_max(self, t):
        
        return max(self.cell_list[:, t])
    
    
    
    
        
        
        
        
    
    
            
            
        
```

```python
class people:
    def __init__(self, count, who):
        
        self.persons=[]
        self.count=count
        self.who=who
        
        for i in range(0, count):
            self.persons.append(Person(who))
        self.persons=np.array(self.persons)
    
        
    def get_mean(self, t, k): 
        return self.persons[k].get_mean(t)

    def get_min(self,t, k):
        return self.persons[k].get_min(t)
    
    def get_max(self, t, k):
        return self.persons[k].get_max(t)
    
    def get_a_min(self, t, ex=-1):
        y_min=[self.get_min(t, k) for k in range(0, self.count) if not k==ex]
        return min(y_min)
    
    def get_a_max(self, t, ex=-1):
        y_min=[self.get_min(t,k) for k in range(0, self.count) if not k==ex]
        return max(y_min)
    
    def get_b_min(self, t, ex=-1):
        y_max=[self.get_max(t, k) for k in range(0, self.count) if not k==ex]
        return min(y_max)
    
    def get_b_max(self, t, ex=-1):
        y_max=[self.get_max(t, k) for k in range(0, self.count) if not k==ex]
        return max(y_max)
    
    def get_c_min(self, t, ex=-1):
        y_mean=[self.get_mean(t, k) for k in range(0, self.count) if not k==ex]
        return min(y_mean)
    
    def get_c_max(self, t, ex=-1):
        y_mean=[self.get_mean(t, k) for k in range(0, self.count) if not k==ex]
        return max(y_mean)
    
    def in_a(self, x, t, ex=-1):
        f1=(x>=self.get_a_min(t, ex))
        f2=(x<=self.get_a_max(t, ex))
        if f1 and f2:
            return 1
        else:
            return 0
    
    def in_b(self, x, t, ex=-1):
        f1=(x>=self.get_b_min(t, ex))
        f2=(x<=self.get_b_max(t, ex))
        if f1 and f2:
            return 1
        else:
            return 0
        
    def in_c(self, x, t, ex=-1):
        f1=(x>=self.get_c_min(t, ex))
        f2=(x<=self.get_c_max(t, ex))
        if f1 and f2:
            return 1
        else:
            return 0
```

```python
def conf_int(h, m, g):
    
    h1=(h*m+g**2/2.0-g*math.sqrt(h*(1-h)*m+g**2/4.0))/(m+g**2)
    h2=(h*m+g**2/2.0+g*math.sqrt(h*(1-h)*m+g**2/4.0))/(m+g**2)
    
    return (h1, h2)
```

```python
def p_stat(x, y):
    x=np.sort(x)
    y=np.sort(y)
    
    N=x.size*(x.size-1)/2.0
    L=0
    
    for i in range (0, x.size-1):
        for j in range(i+1, x.size):
            
            pij=(j-i)/float(x.size+1)
            A=0
            
            for k in range(0, y.size):
                
                
                if (x[i]<y[k] and y[k]<x[j]) :
                    A+=1
            
            h=float(A)/y.size
            
            p=conf_int(h, N, 3)
            
            if pij>p[0] and pij<p[1] :
                if h!=0:
                    L+=1
                else:
                    p=np.random.binomial(1,0.7)
                    L+=p
    
    return float(L)/N
```

```python
class line():
    def __init__(self, mass, point=True):
        if point:
            self.A=mass[1][1]-mass[0][1]
            self.B=mass[0][0]-mass[1][0]
            self.C=mass[0][1]*(mass[1][0]-mass[0][0])-mass[0][0]*(mass[1][1]-mass[0][1])
        else:
            self.A=mass[0]
            self.B=mass[1]
            self.C=mass[2]
            
    def funk1(self, p1,p2):
        self.A=p2[1]-p1[1]
        self.B=p1[0]-p2[0]
        self.C=p1[1]*(p2[0]-p1[0])-p1[0]*(p2[1]-p1[1])
    def eval_y(self, x):
        return (-self.A*x-self.C)/self.B
    
    def evalv(self, p):
        return self.A*p[0]+self.B*p[1]+self.C
    
    def lplot(self, a, b):
        x=np.arange(a, b, 0.01)
        y=[]
        for i in x:
            y.append(self.eval_y(i))
        plt.plot(x, y)
  
        
        
    def intersect(self,l):
        x= self.B*l.C-l.B*self.C
        y=l.A*self.C-self.A*l.C
        w=self.A*l.B-l.A*self.B
        x=x/w
        y=y/w
        return np.array([x, y])
    
    def dist(self,p):
        return (self.A*p[0]+self.B*p[1]+self.C)/math.sqrt(self.A**2+self.B**2)
    
    def on_one_side(self, point1, point2):
        if self.evalv(point1)*self.evalv(point2)>=0:
            return True
        else :
            return False
        
        
```

```python
def two_max(l,M):
    d1=0
    d2=0
    ind1=-1
    ind2=-1
    for i in range (0, len(M)):
        if l.evalv(M[i])>=0:
            if l.dist(M[i])>d1:
                ind1=i
                d1=l.dist(M[i])
        if l.evalv(M[i])<=0:
            if abs (l.dist(M[i]))>d2:
                ind2=i
                d2=abs(l.dist(M[i]))
    return [ind1, ind2]
            
```

```python
def left_down(intersect):
    ind=-1
    mind=min([x[1] for x in intersect])
    ind=[x[0] for x in zip (range(0, len(intersect)), intersect) if x [1][1]==mind]
    if len(ind)>1:
        if intersect[ind[0]][0]>intersect[ind[1]][0]:
            ind = ind[1]
        else:
            ind=ind[0]
    else:
        ind=ind[0]
    return ind
```

```python
def shift(M, p):
    for i in range(0, len(M)):
        M[i][0]-=p[0]
        M[i][1]-=p[1]
  
    return M
```

```python
def neigh(M, ind):
    max_d=0
    ind1=-1
    for i in range (0, len(M)):
        if np.linalg.norm(M[i]- M[ind])>max_d:
            max_d=np.linalg.norm(M[i]-M[ind])
            ind1=i
    res=[]
    for i in range (0, len(M)):
        if i!=ind and i!=ind1:
            res.append(i)
    return res
```

```python
def rotating(M, p1, p2):
    p1=p1/np.linalg.norm(p1)
    p2=p2/np.linalg.norm(p2)
    a=np.zeros((2, 2))
    a[0, 0]=p1[0]
    a[1, 0]=p1[1]
    a[0, 1]=p2[0]
    a[1, 1]=p2[1]
    a_inv=np.linalg.inv(a)
    for i in range (0, len(M)):
        M[i]=np.dot(a_inv, M[i])
    return [M, a]
```

```python
def compr(M,iint,  left_d, neib):
    a=np.linalg.norm(left_d-neib[0])
    b=np.linalg.norm(left_d-neib[1])
    if neib[0][0]==0:
        alph=a/b
    else:
        alph=b/a
   
    for i in range (0, len(M)):
        M[i][0]*=alph
    for i in range (0, len(iint)):
        iint[i][0]*=alph
    return M, iint, alph
    
```

```python
def find_r(M, ct):
    r=0
    for i in range (0, len(M)):
        if np.linalg.norm(ct-M[i])>r:
            r=np.linalg.norm(ct-M[i])
    return r
    
```

```python
def A1(point_ar, E_ar, t, s):
    return E_ar[t][s].in_elips(point_ar[t][s].get_point())

def A3 (point_ar, E_ar, E_two_ar, t, s):
    f1=E_ar[t][s].in_elips(point_ar[t][s].get_point())
    f2=E_two_ar[t][s].in_elips(point_ar[t][s].get_point())
    
    return (f1 and not f2) 

def B1(point_ar, E_ar,E_two_ar, t, s):
    ct1=E_ar[t][s].ct
    ct2=E_two_ar[t][s].ct
    a=ct2[1]-ct1[1]
    b=ct1[0]-ct2[0]
    l_mperp=line([-b, a, 0], False)
    mean_point=(ct1+ct2)/2
    c=l_mperp.evalv(mean_point)
    l_mperp.C=-c
    if l_mperp.on_one_side(point_ar[t][s].get_point(),E_ar[t][s].ct ):
        return True
    else:
        return False
    
def C1(point_ar_p, point_ar_d, E_arp, E_two_arp, E_ard, E_two_ard, t, s ):
    
    f1=A3(point_ar_p, E_arp, E_two_arp, t, s)
    f2=A3(point_ar_d, E_two_ard, E_ard, t, s)
        
    return (f1 or f2)

def C2(point_ar_p, point_ar_d, E_arp, E_two_arp, E_ard, E_two_ard, t, s ):
    f1=A3(point_ar_p, E_two_arp, E_arp, t, s)
    f2=A3(point_ar_d, E_ard, E_two_ard, t, s)
    return (f1 or f2)

def C3(point_ar_p, point_ar_d, E_arp,E_two_ard, t, s):
    f1=A1(point_ar_p, E_arp, t, s)
    f2=A1(point_ar_d, E_two_ard, t, s)
    return (f1 or f2)

def C4(point_ar_p, point_ar_d, E_two_arp,E_ard, t, s):
    f1=A1(point_ar_p, E_two_arp, t, s)
    f2=A1(point_ar_d, E_ard, t, s)
    return (f1 or f2)

def C5(point_ar_p, point_ar_d, E_arp, E_two_arp, E_ard, E_two_ard, t, s ):
    f1=B1(point_ar_p, E_arp,E_two_arp, t, s)
    f2=B1(point_ar_d, E_two_ard,E_ard, t, s)
    return (f1 or f2)


def C6 (point_ar_p, point_ar_d, E_arp, E_two_arp, E_ard, E_two_ard, t, s ):
    f1=B1(point_ar_p, E_two_arp,E_arp, t, s)
    f2=B1(point_ar_d, E_ard,E_two_ard, t, s)
    return (f1 or f2)
```

```python
def find_alpha(pep1, pep2, count_feat, who):
    if who=='rmg':
        pep=pep1
    if who=='fam':
        pep=pep2
    alpha=np.zeros((len(pep.persons), 2))
    for ex in range(0, len(pep.persons)):
  
        for t in range (0, count_feat):
            x_mean_t=pep.get_mean( t, ex )
            x_min_t=pep.get_min( t, ex)
            x_max_t=pep.get_max( t, ex)
            alpha[ex, 0]+=pep1.in_a(x_min_t, t, ex)
            alpha[ex,0 ]+=pep1.in_c(x_mean_t, t, ex)
            alpha[ex, 0]+=pep1.in_b(x_max_t, t, ex)
            alpha[ex, 1]+=pep2.in_a(x_min_t, t)
            alpha[ex, 1]+=pep2.in_b(x_max_t, t)
            alpha[ex, 1]+=pep2.in_c(x_mean_t, t)
    return alpha

    
```
