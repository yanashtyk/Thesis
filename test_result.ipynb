```python
import numpy as np
```

```python
def sq (h):
    if h[2]>h[3]:
        return 1
    else:
        return 0

def lin(h):
    if h[4]>h[5]:
        return 1
    else:
        return 0
    
```

```python
def find_sq(h):
    count_rmg=0
    for i in range(0, h.shape[0]):
        if sq(h[i])==1:
            count_rmg+=1
    return float(count_rmg)/h.shape[0]


def find_lin(h):
    count_rmg=0
    for i in range (0, h.shape[0]):
        if lin(h[i])==1:
            count_rmg+=1
    return float(count_rmg)/h.shape[0]
```

```python
def tot_f(h):
    if h[0]>h[1]:
        return 0
    else:
        return 1
```

```python
def tot(h1):
    count=0
    for i in range(0,h1.shape[0]):
        count +=tot_f(h1[i, :])
        
    return (count / float(h1.shape[0]), 1-count/float(h1.shape[0]))
        
```

```python
def sus_dom(h):
    count_sd_fam=0
    count_sd_rmg=0
    for i in range(0, h.shape[0]):
        if h[i][3]>=h[i][2] and h[i][1]>=h[i][0]:
            count_sd_fam+=1
        if h[i][3]<=h[i][2] and h[i][1]<=h[i][0]:
            count_sd_rmg+=1
    return (count_sd_fam/float(h.shape[0]), count_sd_rmg/h.shape[0])
```
